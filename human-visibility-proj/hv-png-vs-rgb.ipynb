```python
from requests_html import HTMLSession
import wget
```

```python
def extract_imgid(urls_list):
    assert isinstance(urls_list, list), "Expected a list"
    dict_temp = {}
    for v in urls_list:
        img_name = v.split('/')[-1]
        k = int(''.join([c for c in img_name if c.isdigit()]))
        dict_temp[k] = v
    return dict_temp
```

#### Find image diffs PNG vs RAW.

```python
for sex in ('Female', 'Male'):

    #### Get top-level index URLs
    
    rawz_url = f'https://data.lhncbc.nlm.nih.gov/public/Visible-Human/{sex}-Images/Fullcolor/fullbody/index.html'
    png_url = f'https://data.lhncbc.nlm.nih.gov/public/Visible-Human/{sex}-Images/PNG_format/index.html'
    
    session = HTMLSession()
    
    #### Build image URLs map.
    
    # Get raw.Z URLs
    r = session.get(rawz_url)
    raw_list = list(r.html.absolute_links)
    urls_raw = extract_imgid(raw_list) # Dictionary with img_id, img_url
    
    # Get png URLs
    p = session.get(png_url)
    _temp = list(p.html.absolute_links)
    png_ix = [u for u in _temp if 'radiological' not in u]
    png_ix
    png_list = []
    for url in png_ix:
        r = session.get(url)
        _temp = list(r.html.absolute_links)
        png_list += _temp
    urls_png = extract_imgid(png_list) # Dictionary with img_id, img_url
    
    #### Diffs
    len_raw = len(urls_raw)
    len_png = len(urls_png)
    print(f'{sex}: {len_raw} raw images, {len_png} png images available')
```
