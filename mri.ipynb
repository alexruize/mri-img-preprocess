# Preprocess MRI data in NIfTI format.

Extract, preprocess and stitch together into a file suitable for further post-processing in Touchdesigner or other real-time visual software.

```python
from sys import platform
import os
import nibabel
import matplotlib.pyplot as plt
```

#### System paths

```python
if platform == 'linux':
    homedir = '~'
elif platform == 'win32':
    homedir = '%HOMEDRIVE%%HOMEPATH%'
else:
    raise ValueError("Cannot set home directory for this OS.")
```

#### Data directory

```python
datadir = r'Downloads\ds004767-download'
```

#### Working directory

```python
wk_dir = os.path.expandvars(os.path.join(homedir, datadir))
wk_dir
```

#### Open NiFTI dataset

```python
f_nifti = r'sub-00xHNL21\anat\sub-00xHNL21_T2w.nii\HNL-11-15_hippo_200um_stitch_20171017.nii'
```

```python
nifti_path = os.path.join(wk_dir, f_nifti)
nifti_data = nibabel.load(nifti_path)
mri = nifti_data.get_fdata() # Creates numpy.memmap object (lazy loading)
```

```python
img = mri[:, :, 80]
img 
```
